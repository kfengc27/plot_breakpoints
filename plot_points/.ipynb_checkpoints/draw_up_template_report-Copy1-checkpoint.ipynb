```python
import numpy as np
import matplotlib.pyplot as plt

def monotone_nearest_path(x, y, allow_equal_y=True, max_dx=None, max_dist=None):
    x = np.asarray(x); y = np.asarray(y)
    idx = np.argsort(x)                 # 先按 x 排序
    xs, ys = x[idx], y[idx]

    path_idx = [0]                      # 从最左点开始
    curr = 0

    while True:
        x0, y0 = xs[curr], ys[curr]

        # 候选：在右侧、且不下降
        j = np.arange(curr+1, len(xs))
        mask = (xs[j] > x0)
        if allow_equal_y:
            mask &= (ys[j] >= y0)
        else:
            mask &= (ys[j] > y0)

        if max_dx is not None:
            mask &= (xs[j] - x0) <= max_dx

        cand = j[mask]
        if cand.size == 0:
            break

        dx = xs[cand] - x0
        dy = ys[cand] - y0
        dist = np.hypot(dx, dy)

        if max_dist is not None:
            ok = dist <= max_dist
            cand, dist = cand[ok], dist[ok]
            if cand.size == 0:
                break

        # 选距离最小的
        next_j = cand[np.argmin(dist)]

        path_idx.append(next_j)
        curr = next_j

    return xs[path_idx], ys[path_idx], idx[path_idx]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def keep_more_points(x, y, x_path, y_path,
                     abs_tol=100.0,      # 绝对容忍（↑更宽）
                     rel_tol=0.012,      # 相对容忍（按路径高度比例；↑更宽）
                     k_mad=2.0,          # 局部稳健波动权重（↑更宽）
                     q_low=15,           # 分箱低分位兜底（↓更宽；如 10）
                     jump_min=60.0,      # 识别“台阶跳升”的阈值
                     jump_bonus=80.0,    # 台阶附近额外放宽（↑更宽）
                     x_radius_frac=0.03, # 台阶影响半径（占 x 范围比例；↑更宽）
                     max_drop_cap=220.0  # 允许向下最大下探（↑更宽）
                    ):
    x = np.asarray(x); y = np.asarray(y)
    x_path = np.asarray(x_path); y_path = np.asarray(y_path)

    # 路径按 x 递增
    order = np.argsort(x_path)
    x_path = x_path[order]; y_path = y_path[order]

    # 路径插值（裁剪到路径范围，避免端点水平外推）
    x_min, x_max = x_path[0], x_path[-1]
    x_clip = np.clip(x, x_min, x_max)
    y_on_path = np.interp(x_clip, x_path, y_path)

    # 残差：点相对路径的高度（负=在路径下方）
    r = y - y_on_path

    # 分箱做局部统计
    N = len(x)
    n_bins = max(20, min(120, N // 150))
    edges = np.linspace(x.min(), x.max(), n_bins + 1)
    bin_id = np.digitize(x, edges) - 1
    bin_id = np.clip(bin_id, 0, n_bins - 1)

    # 基础阈值（负数，越负越宽）
    thr = -(abs_tol + rel_tol * y_on_path) * np.ones_like(y, float)

    # 叠加 MAD 稳健尺度
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15: 
            continue
        r_b = r[idx]
        med = np.median(r_b)
        mad = np.median(np.abs(r_b - med))
        sigma = 1.4826 * mad
        thr[idx] -= k_mad * sigma

    # 分箱低分位兜底（确保每箱都留一批“靠近平台”的点）
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        thr_q = np.percentile(r_b, q_low)
        thr[idx] = np.minimum(thr[idx], thr_q)

    # 台阶跳升附近额外放宽（照顾新平台略低于红线的簇）
    dy = np.diff(y_path)
    jump_idx = np.where(dy >= jump_min)[0] + 1
    jump_x = x_path[jump_idx]
    x_radius = x_radius_frac * (x.max() - x.min())

    bonus = np.zeros_like(y, float)
    for jx in jump_x:
        near = np.abs(x - jx) <= x_radius
        bonus[near] += jump_bonus
    thr -= bonus

    # 最大下探保护
    thr = np.maximum(thr, -max_drop_cap)

    mask_keep = r >= thr
    return mask_keep, y_on_path

# ===== 直接使用（推荐 Balanced 档）=====
# 三个预设：
presets = {
    "Conservative": dict(abs_tol=60,  rel_tol=0.008, k_mad=1.5, q_low=20, jump_min=60, jump_bonus=40, x_radius_frac=0.025, max_drop_cap=160),
    "Balanced":     dict(abs_tol=100, rel_tol=0.012, k_mad=2.0, q_low=15, jump_min=60, jump_bonus=80, x_radius_frac=0.03,  max_drop_cap=220),
    "Generous":     dict(abs_tol=140, rel_tol=0.016, k_mad=2.5, q_low=10, jump_min=50, jump_bonus=100,x_radius_frac=0.035, max_drop_cap=280),
}

mask_keep, y_on_path = keep_more_points(x, y, x_path, y_path, **presets["Balanced"])

# 可视化（区分 kept / dropped）
idx = np.argsort(x)
plt.figure(figsize=(12,6))
plt.scatter(np.array(x)[~mask_keep], np.array(y)[~mask_keep], s=6, alpha=0.15, marker='x', label='dropped')
plt.scatter(np.array(x)[mask_keep],  np.array(y)[mask_keep],  s=8, alpha=0.35,              label='kept (near path)')
# plt.plot(np.array(x_path)[np.argsort(x_path)], np.array(y_path)[np.argsort(y_path)], lw=1.5, label='nearest upward path')
plt.xlabel('accumulative complexity'); plt.ylabel('complexity')
plt.title('Strict nearest-neighbor upward path (filtered, keep more)')
plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ========= 1) 你的原函数：保留“黄色点” =========
def keep_more_points(x, y, x_path, y_path,
                     abs_tol=100.0,      # 绝对容忍（↑更宽）
                     rel_tol=0.012,      # 相对容忍（按路径高度比例；↑更宽）
                     k_mad=2.0,          # 局部稳健波动权重（↑更宽）
                     q_low=12,           # 分箱低分位兜底（↓更宽；如 10）
                     jump_min=60.0,      # 识别“台阶跳升”的阈值
                     jump_bonus=80.0,    # 台阶附近额外放宽（↑更宽）
                     x_radius_frac=0.03, # 台阶影响半径（占 x 范围比例；↑更宽）
                     max_drop_cap=220.0  # 允许向下最大下探（↑更宽）
                    ):
    x = np.asarray(x); y = np.asarray(y)
    x_path = np.asarray(x_path); y_path = np.asarray(y_path)

    # 路径按 x 递增
    order = np.argsort(x_path)
    x_path = x_path[order]; y_path = y_path[order]

    # 路径插值（裁剪到路径范围，避免端点水平外推）
    x_min, x_max = x_path[0], x_path[-1]
    x_clip = np.clip(x, x_min, x_max)
    y_on_path = np.interp(x_clip, x_path, y_path)

    # 残差：点相对路径的高度（负=在路径下方）
    r = y - y_on_path

    # 分箱做局部统计
    N = len(x)
    n_bins = max(20, min(120, N // 150))
    edges = np.linspace(x.min(), x.max(), n_bins + 1)
    bin_id = np.digitize(x, edges) - 1
    bin_id = np.clip(bin_id, 0, n_bins - 1)

    # 基础阈值（负数，越负越宽）
    thr = -(abs_tol + rel_tol * y_on_path) * np.ones_like(y, float)

    # 叠加 MAD 稳健尺度
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15: 
            continue
        r_b = r[idx]
        med = np.median(r_b)
        mad = np.median(np.abs(r_b - med))
        sigma = 1.4826 * mad
        thr[idx] -= k_mad * sigma

    # 分箱低分位兜底（确保每箱都留一批“靠近平台”的点）
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        thr_q = np.percentile(r_b, q_low)
        thr[idx] = np.minimum(thr[idx], thr_q)

    # 台阶跳升附近额外放宽（照顾新平台略低于红线的簇）
    dy = np.diff(y_path)
    jump_idx = np.where(dy >= jump_min)[0] + 1
    jump_x = x_path[jump_idx]
    x_radius = x_radius_frac * (x.max() - x.min())

    bonus = np.zeros_like(y, float)
    for jx in jump_x:
        near = np.abs(x - jx) <= x_radius
        bonus[near] += jump_bonus
    thr -= bonus

    # 最大下探保护
    thr = np.maximum(thr, -max_drop_cap)

    mask_keep = r >= thr
    return mask_keep, y_on_path

# ========= 2) 递归删除“下面的点”（批量递归版：快） =========
def prune_downward_recursive(x, y, drop_tol=50.0, max_depth=2):
    """
    对 (x,y) 批量递归删除“明显在当前平台之下”的点：
        从左到右扫描，保持当前最大 y_max；
        若 y_i + drop_tol < y_max，则此点判为“下降”，本轮剔除；
        若本轮有剔除，递归下一轮，直到收敛或达最大深度。
    返回：
        x_clean, y_clean, kept_pos_in_input（在输入序列里的位置索引，按清洗后的顺序）
    说明：
        - 输入不要求已按 x 排序；函数内部会按 x 升序处理。
        - drop_tol 越小越严格（0~30 常用；平台抖动大时可取 20~40）。
    """
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x)
    xs, ys = x[order], y[order]

    def one_pass(xs, ys):
        keep = np.zeros_like(xs, dtype=bool)
        y_max = -np.inf
        for i in range(len(xs)):
            if ys[i] + drop_tol >= y_max:
                keep[i] = True
                if ys[i] > y_max:
                    y_max = ys[i]
        return keep

    depth = 0
    kept_mask = np.ones_like(xs, dtype=bool)
    while depth <= max_depth:
        keep = one_pass(xs, ys)
        if keep.all():
            break
        xs, ys = xs[keep], ys[keep]
        depth += 1

    if depth > max_depth:
        raise RecursionError("Exceeded max recursion depth")

    # 把清洗后点的位置映射回“排序后的输入”位置
    # 简单匹配（若有完全重复点，会匹配到先出现的一个）
    kept_positions_sorted = []
    j = 0
    for i in range(len(np.sort(x))):
        if j < len(xs) and order[i] < len(x) and x[order[i]] == xs[j] and y[order[i]] == ys[j]:
            kept_positions_sorted.append(i)
            j += 1
            if j == len(xs):
                break

    kept_positions_sorted = np.array(kept_positions_sorted, dtype=int)
    kept_pos_in_input = order[kept_positions_sorted]  # 相对于输入 (x,y) 的索引
    return xs, ys, kept_pos_in_input

# ========= 3) 预设集合 =========
presets = {
    "Conservative": dict(abs_tol=60,  rel_tol=0.008, k_mad=1.5, q_low=20, jump_min=60, jump_bonus=40, x_radius_frac=0.025, max_drop_cap=160),
    "Balanced":     dict(abs_tol=100, rel_tol=0.012, k_mad=2.0, q_low=15, jump_min=60, jump_bonus=80, x_radius_frac=0.03,  max_drop_cap=220),
    "Generous":     dict(abs_tol=140, rel_tol=0.016, k_mad=2.5, q_low=10, jump_min=50, jump_bonus=100,x_radius_frac=0.035, max_drop_cap=280),
}

# ========= 4) —— 使用流程 —— =========
# 先得到黄色点（mask_keep）
mask_keep, y_on_path = keep_more_points(x, y, x_path, y_path, **presets["Balanced"])

# 黄色点子集
x_keep = np.array(x)[mask_keep]
y_keep = np.array(y)[mask_keep]

# 递归批量清理“下面的点”（把 drop_tol 调小更严格）
x_clean, y_clean, kept_pos_within_keep = prune_downward_recursive(x_keep, y_keep, drop_tol=90.0)

# 如果你需要“清理后点”在原始数组中的索引：
kept_idx_full = np.nonzero(mask_keep)[0][kept_pos_within_keep]  # 相对于原始 (x,y)



# ========= 5) 画图对比 =========
plt.figure(figsize=(12,6))
plt.scatter(np.array(x)[~mask_keep], np.array(y)[~mask_keep], s=6, alpha=0.12, marker='x', label='dropped (path filter)')

# 递归前的黄色点（浅黄）
plt.scatter(x_keep, y_keep, s=10, alpha=0.35, color='green', edgecolors='none', label='kept (before recursion)')

# 递归后的点（高亮）
plt.scatter(x_clean, y_clean, s=14, alpha=0.9, color='orange', edgecolors='black', label='kept (after recursive prune)')



# 可选：画路径
# order_p = np.argsort(x_path)
# plt.plot(np.array(x_path)[order_p], np.array(y_path)[order_p], lw=1.2, label='path')

plt.xlabel('accumulative complexity'); plt.ylabel('complexity')
plt.title('Keep-more + Recursive pruning of downward points')
plt.legend(); plt.tight_layout(); plt.show()
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt

# ========== 1) 第一阶段：保留“黄色点”（返回 thr 便于画阈值线） ==========
def keep_more_points(x, y, x_path, y_path,
                     abs_tol=100.0, rel_tol=0.012, k_mad=2.0,
                     q_low=12, jump_min=60.0, jump_bonus=80.0,
                     x_radius_frac=0.03, max_drop_cap=220.0):
    x = np.asarray(x); y = np.asarray(y)
    x_path = np.asarray(x_path); y_path = np.asarray(y_path)

    order = np.argsort(x_path)
    x_path = x_path[order]; y_path = y_path[order]

    x_min, x_max = x_path[0], x_path[-1]
    x_clip = np.clip(x, x_min, x_max)
    y_on_path = np.interp(x_clip, x_path, y_path)

    r = y - y_on_path

    N = len(x)
    n_bins = max(20, min(120, N // 150))
    edges = np.linspace(x.min(), x.max(), n_bins + 1)
    bin_id = np.clip(np.digitize(x, edges) - 1, 0, n_bins - 1)

    thr = -(abs_tol + rel_tol * y_on_path) * np.ones_like(y, float)

    # 叠加 MAD 稳健尺度
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        med = np.median(r_b)
        mad = np.median(np.abs(r_b - med))
        sigma = 1.4826 * mad
        thr[idx] -= k_mad * sigma

    # 分箱低分位兜底
    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        thr_q = np.percentile(r_b, q_low)
        thr[idx] = np.minimum(thr[idx], thr_q)

    # 台阶跳升附近额外放宽
    dy = np.diff(y_path)
    jump_idx = np.where(dy >= jump_min)[0] + 1
    jump_x = x_path[jump_idx]
    x_radius = x_radius_frac * (x.max() - x.min())

    bonus = np.zeros_like(y, float)
    for jx in jump_x:
        near = np.abs(x - jx) <= x_radius
        bonus[near] += jump_bonus
    thr -= bonus

    # 最大下探保护
    thr = np.maximum(thr, -max_drop_cap)

    mask_keep = r >= thr
    return mask_keep, y_on_path, thr

# ========== 2) 第二阶段：递归清理（偏向保留 Y 更高的点） ==========
def prune_downward_recursive_keep_high(
    x, y,
    drop_tol=110.0,   # 越大越宽松（80~140 常用）
    protect_q=0.90,   # 局部高分位保护（0.85~0.92）
    win=401,          # 滚动窗口（奇数）
    patience=3,       # 连续违例数（越大越宽）
    max_rounds=12
):
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x)
    xs, ys = x[order], y[order]
    idx_sorted = order.copy()

    def rolling_quantile(a, q, w):
        n, h = len(a), w // 2
        ap = np.concatenate([np.full(h, a[0]), a, np.full(h, a[-1])])
        out = np.empty(n)
        for i in range(n):
            out[i] = np.quantile(ap[i:i+w], q)
        return out

    rounds = 0
    while rounds < max_rounds and len(ys) > 0:
        q_hi = rolling_quantile(ys, protect_q, win)
        is_high = ys >= q_hi

        keep = np.ones(len(ys), dtype=bool)
        run_bad = 0
        y_max = -np.inf

        for i in range(len(ys)):
            v = ys[i]

            # 高点直接保留，并更新平台
            if is_high[i]:
                keep[i] = True
                if v > y_max:
                    y_max = v
                run_bad = 0
                continue

            # 非高点：不得明显低于当前平台
            if v + drop_tol < y_max:
                run_bad += 1
            else:
                run_bad = 0
                if v > y_max:
                    y_max = v

            if run_bad >= patience:
                keep[i] = False

        if keep.all():
            break

        xs, ys = xs[keep], ys[keep]
        idx_sorted = idx_sorted[keep]
        rounds += 1

    kept_idx_in_input = idx_sorted
    return xs, ys, kept_idx_in_input

# ========== 3) 合并：优先用递归结果，否则用原始 (x,y) ==========
def merge_preferring_recursive(
    x, y,
    kept_idx_full=None, y_clean=None,
    x_clean=None, y_clean_by_x=None, x_match_tol=0.0
):
    """
    返回：X_out, Y_out（按 x 升序）
    规则：从左到右；若该位置在“递归后结果”中，用递归后的 Y；否则用原始 y。
    优先使用 kept_idx_full（索引方式，最稳）；否则用 (x_clean, y_clean_by_x) 按 x 匹配。
    """
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x)
    X_out = x[order].copy()
    Y_out = y[order].copy()

    # 方式 A：用原始索引对齐（推荐）
    if kept_idx_full is not None:
        kept_idx_full = np.asarray(kept_idx_full, dtype=int)
        if y_clean is None:
            y_clean = y[kept_idx_full]
        else:
            y_clean = np.asarray(y_clean)

        pos_in_sorted = np.empty_like(order)
        pos_in_sorted[order] = np.arange(len(order))
        positions = pos_in_sorted[kept_idx_full]
        Y_out[positions] = y_clean
        return X_out, Y_out

    # 方式 B：只有 (x_clean, y_clean)，用 x 匹配
    if x_clean is not None and y_clean_by_x is not None:
        xr = np.asarray(x_clean); yr = np.asarray(y_clean_by_x)
        rorder = np.argsort(xr)
        xr, yr = xr[rorder], yr[rorder]
        j = 0
        for i, xi in enumerate(X_out):
            while j < len(xr) and xr[j] < xi - x_match_tol:
                j += 1
            if j < len(xr) and abs(xr[j] - xi) <= x_match_tol:
                Y_out[i] = yr[j]
        return X_out, Y_out

    raise ValueError("请提供 kept_idx_full(+y_clean) 或 x_clean+y_clean_by_x 其中一种。")

# ========== 4) 演示数据（若环境里没有 x,y,x_path,y_path） ==========
def generate_demo_data(n=3500, seed=7):
    rng = np.random.default_rng(seed)
    x = np.sort(rng.uniform(0, 3.5e6, size=n))
    # 构造台阶路径
    brk = [0.4e6, 0.9e6, 1.4e6, 2.1e6, 2.6e6, 3.1e6]
    lev = [1750, 1850, 2000, 2350, 3000, 3350, 3480]
    y_path = np.zeros_like(x)
    last = 0.0
    for i, xv in enumerate(x):
        if xv < brk[0]:
            y_path[i] = lev[0] + 0.02 * xv/1e4
        elif xv < brk[1]:
            y_path[i] = lev[1] + 0.02 * (xv - brk[0])/1e4
        elif xv < brk[2]:
            y_path[i] = lev[2] + 0.04 * (xv - brk[1])/1e4
        elif xv < brk[3]:
            y_path[i] = lev[3] + 0.03 * (xv - brk[2])/1e4
        elif xv < brk[4]:
            y_path[i] = lev[4] + 0.02 * (xv - brk[3])/1e4
        elif xv < brk[5]:
            y_path[i] = lev[5] + 0.015 * (xv - brk[4])/1e4
        else:
            y_path[i] = lev[6] + 0.01 * (xv - brk[5])/1e4
        last = y_path[i]

    y = y_path + rng.normal(0, 35, size=n)
    # 模拟一批下探点
    bad = (x > 0.2e6) & (x < 2.6e6) & (rng.random(n) < 0.12)
    y[bad] -= rng.uniform(60, 220, bad.sum())

    # 稀疏的路径点
    xp = np.linspace(x.min(), x.max(), 90)
    yp = np.interp(xp, x, y_path)
    return x, y, xp, yp

# ========== 5) 主流程 ==========
def main():
    # (A) 取数据：优先使用外部变量
    g = globals()
    if all(k in g for k in ("x", "y", "x_path", "y_path")):
        print(">>> Using existing x, y, x_path, y_path.")
        X, Y, Xp, Yp = np.asarray(g["x"]), np.asarray(g["y"]), np.asarray(g["x_path"]), np.asarray(g["y_path"])
    else:
        print(">>> No external arrays found; generating demo data...")
        X, Y, Xp, Yp = generate_demo_data()

    # (B) 第一阶段：建议用更宽松的参数以保留更多平台边缘点
    presets = dict(abs_tol=140, rel_tol=0.016, k_mad=2.5, q_low=10, jump_min=50, jump_bonus=100, x_radius_frac=0.035, max_drop_cap=280)
    mask_keep, y_on_path, thr = keep_more_points(X, Y, Xp, Yp, **presets)

    X_keep, Y_keep = X[mask_keep], Y[mask_keep]

    # (C) 第二阶段：偏向保留高点的递归清理（宽松设置）
    x_clean, y_clean, kept_pos_within_keep = prune_downward_recursive_keep_high(
        X_keep, Y_keep,
        drop_tol=110.0, protect_q=0.90, win=401, patience=3
    )
    kept_idx_full = np.nonzero(mask_keep)[0][kept_pos_within_keep]  # 映射到原始索引

    # (D) 合并：从左到右，优先用递归后的 y，否则用原始 y
    X_out, Y_out = merge_preferring_recursive(
        X, Y,
        kept_idx_full=kept_idx_full,
        y_clean=y_clean   # 若递归没修改 y，这个也可不传
    )

    # (E) 可视化
    plt.figure(figsize=(14,7))
    plt.scatter(X[~mask_keep], Y[~mask_keep], s=6, alpha=0.08, color='skyblue', label='dropped (stage-1)')
    plt.scatter(X_keep, Y_keep, s=10, alpha=0.35, color='red', label='kept (stage-1)')
    plt.scatter(x_clean, y_clean, s=16, alpha=0.9, color='gold', edgecolors='black', label='kept (after recursion)')
    # 合并后的“左→右优先递归”曲线
    # plt.plot(X_out, Y_out, lw=1.2, color='black', alpha=0.9, label='merged (prefer recursive)')
    # 路径与阈值（可选）
    op = np.argsort(Xp)
    plt.plot(Xp[op], Yp[op], lw=1.1, color='tomato', alpha=0.9, label='path')
    ord_all = np.argsort(X)
    plt.plot(X[ord_all], (y_on_path + thr)[ord_all], lw=1.0, ls='--', alpha=0.9, label='keep threshold')
    plt.xlabel('accumulative complexity'); plt.ylabel('complexity')
    plt.title('Stage-1 keep → High-preserving recursion → Left-to-right merge (prefer recursive)')
    plt.legend(); plt.tight_layout(); plt.show()

    print(f"Stage-1 kept: {mask_keep.sum()} / {len(X)} ({mask_keep.mean():.1%})")
    print(f"Recursive kept: {len(x_clean)} / {len(X_keep)} ({len(x_clean)/len(X_keep):.1%})")
    print(f"Merged series length: {len(X_out)} (same as raw X)")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ========= 1) 保留“黄色点”（返回 thr 便于画阈值线） =========
def keep_more_points(x, y, x_path, y_path,
                     abs_tol=100.0,
                     rel_tol=0.012,
                     k_mad=2.0,
                     q_low=2,          # 你设为 12：已比 15 更宽松
                     jump_min=60.0,
                     jump_bonus=80.0,
                     x_radius_frac=0.03,
                     max_drop_cap=220.0):
    x = np.asarray(x); y = np.asarray(y)
    x_path = np.asarray(x_path); y_path = np.asarray(y_path)

    order = np.argsort(x_path)
    x_path = x_path[order]; y_path = y_path[order]

    x_min, x_max = x_path[0], x_path[-1]
    x_clip = np.clip(x, x_min, x_max)
    y_on_path = np.interp(x_clip, x_path, y_path)

    r = y - y_on_path

    N = len(x)
    n_bins = max(20, min(120, N // 150))
    edges = np.linspace(x.min(), x.max(), n_bins + 1)
    bin_id = np.digitize(x, edges) - 1
    bin_id = np.clip(bin_id, 0, n_bins - 1)

    thr = -(abs_tol + rel_tol * y_on_path) * np.ones_like(y, float)

    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        med = np.median(r_b)
        mad = np.median(np.abs(r_b - med))
        sigma = 1.4826 * mad
        thr[idx] -= k_mad * sigma

    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        thr_q = np.percentile(r_b, q_low)
        thr[idx] = np.minimum(thr[idx], thr_q)

    dy = np.diff(y_path)
    jump_idx = np.where(dy >= jump_min)[0] + 1
    jump_x = x_path[jump_idx]
    x_radius = x_radius_frac * (x.max() - x.min())

    bonus = np.zeros_like(y, float)
    for jx in jump_x:
        near = np.abs(x - jx) <= x_radius
        bonus[near] += jump_bonus
    thr -= bonus

    thr = np.maximum(thr, -max_drop_cap)

    mask_keep = r >= thr
    return mask_keep, y_on_path, thr   # ← 新增 thr

# ========= 2) 递归删除“下面的点”（你的版本不变） =========
def prune_downward_recursive(x, y, drop_tol=1.0, max_depth=1000):
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x)
    xs, ys = x[order], y[order]

    def one_pass(xs, ys):
        keep = np.zeros_like(xs, dtype=bool)
        y_max = -np.inf
        for i in range(len(xs)):
            if ys[i] + drop_tol >= y_max:
                keep[i] = True
                if ys[i] > y_max:
                    y_max = ys[i]
        return keep

    depth = 0
    while depth <= max_depth:
        keep = one_pass(xs, ys)
        if keep.all():
            break
        xs, ys = xs[keep], ys[keep]
        depth += 1

    if depth > max_depth:
        raise RecursionError("Exceeded max recursion depth")

    kept_positions_sorted = []
    j = 0
    for i in range(len(order)):
        if j < len(xs) and x[order[i]] == xs[j] and y[order[i]] == ys[j]:
            kept_positions_sorted.append(i)
            j += 1
            if j == len(xs):
                break

    kept_pos_in_input = order[np.array(kept_positions_sorted, dtype=int)]
    return xs, ys, kept_pos_in_input

# ========= 3) 预设 =========
presets = {
    "Conservative": dict(abs_tol=60,  rel_tol=0.008, k_mad=1.5, q_low=20, jump_min=60, jump_bonus=40, x_radius_frac=0.025, max_drop_cap=160),
    "Balanced":     dict(abs_tol=100, rel_tol=0.012, k_mad=2.0, q_low=15, jump_min=60, jump_bonus=80, x_radius_frac=0.03,  max_drop_cap=220),
    "Generous":     dict(abs_tol=140, rel_tol=0.016, k_mad=2.5, q_low=10, jump_min=50, jump_bonus=100,x_radius_frac=0.035, max_drop_cap=280),
}

# ========= 4) 使用流程 =========
# 第一阶段：建议先更宽松一些（想保留红箭头那批）
mask_keep, y_on_path, thr = keep_more_points(x, y, x_path, y_path, **presets["Generous"])

x = np.asarray(x); y = np.asarray(y)
x_keep = x[mask_keep]; y_keep = y[mask_keep]

# 第二阶段：把 drop_tol 调大 = 更宽松（例如 80~120）
x_clean, y_clean, kept_pos_within_keep = prune_downward_recursive(x_keep, y_keep, drop_tol=90.0)

# 若需要最终索引（相对原始 (x,y)）
kept_idx_full = np.nonzero(mask_keep)[0][kept_pos_within_keep]

# ========= 5) 画图（含路径与阈值线） =========
plt.figure(figsize=(13,6))

# 原始
plt.scatter(x, y, s=6, alpha=0.07, color='gray', label='all raw')

# 阶段1（黄色）
plt.scatter(x_keep, y_keep, s=10, alpha=0.35, color='gold', edgecolors='none', label='kept (stage-1)')

# 阶段2（橙色+黑边）
plt.scatter(x_clean, y_clean, s=16, alpha=0.9, color='orange', edgecolors='black', label='after recursion')

# 路径
op = np.argsort(x_path)
plt.plot(np.asarray(x_path)[op], np.asarray(y_path)[op], lw=1.2, color='tomato', alpha=0.9, label='path')

# 阈值线：在 x 全体上画 path + thr（thr 对应 y_on_path 的 x）
ord_all = np.argsort(x)
plt.plot(x[ord_all], (y_on_path + thr)[ord_all], lw=1.0, ls='--', alpha=0.9, label='keep threshold')

plt.xlabel('accumulative complexity'); plt.ylabel('complexity')
plt.title('Stage-1 keep (with threshold) → recursive prune (lenient)')
plt.legend()
plt.tight_layout(); plt.show()

```

```python
import numpy as np

def prune_downward_recursive_v2(x, y,
                                q_hi=0.80,        # 局部“上平台”用的高分位（0.75~0.9 常用）
                                win=401,          # 滚动窗口（奇数），越大越平滑
                                abs_tol=40.0,     # 绝对下探容忍
                                rel_tol=0.010,    # 相对下探容忍，相对于局部平台
                                k_mad=1.5,        # 叠加局部波动（MAD）的宽容
                                patience=3,       # 需连续违例多少个点才删
                                max_rounds=20):   # 递归最多轮数
    """
    温和版递归清理（v2）：
    - 按 x 升序
    - 计算滚动高分位 q_hi 作为局部“平台” (plateau)
    - 计算滚动 MAD 估计局部噪声，形成自适应阈值：floor = plateau - (abs_tol + rel_tol*plateau + k_mad*sigma)
    - 仅当 '连续 patience 个' 低于 floor 的点，才在本轮标为删除
    - 批量删除 -> 进入下一轮，直至没有删除或达到 max_rounds
    返回：
        x_clean, y_clean, kept_idx_in_input（在输入 x,y 中的索引，按清洗后顺序）
    """
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x)
    xs, ys = x[order], y[order]

    # --- 辅助：滚动分位数与 MAD ---
    def rolling_quantile(a, q, w):
        # 简洁稳妥：用 pad + 每步窗口分位（O(n*w)，w 不要太大）
        n = len(a)
        half = w // 2
        pad_left = a[0] * np.ones(half)
        pad_right = a[-1] * np.ones(half)
        ap = np.concatenate([pad_left, a, pad_right])
        out = np.empty(n)
        for i in range(n):
            out[i] = np.quantile(ap[i:i+w], q)
        return out

    def rolling_mad(a, w):
        n = len(a)
        half = w // 2
        pad_left = a[0] * np.ones(half)
        pad_right = a[-1] * np.ones(half)
        ap = np.concatenate([pad_left, a, pad_right])
        out = np.empty(n)
        for i in range(n):
            seg = ap[i:i+w]
            med = np.median(seg)
            mad = np.median(np.abs(seg - med))
            out[i] = 1.4826 * mad   # ~sigma
        return out

    # --- 递归轮 ---
    rounds = 0
    while rounds < max_rounds and len(ys) > 0:
        plateau = rolling_quantile(ys, q_hi, win)
        sigma   = rolling_mad(ys, win)
        floor   = plateau - (abs_tol + rel_tol * np.maximum(1.0, plateau) + k_mad * sigma)

        # 连续违例计数
        keep = np.ones(len(ys), dtype=bool)
        run = 0
        for i in range(len(ys)):
            if ys[i] < floor[i]:
                run += 1
            else:
                run = 0
            if run >= patience:
                keep[i] = False

        if keep.all():
            break  # 收敛

        xs, ys = xs[keep], ys[keep]
        rounds += 1

    # 映射回原始索引（按清洗后顺序）
    kept_positions_sorted = []
    j = 0
    for i in range(len(order)):
        if j < len(xs) and x[order[i]] == xs[j] and y[order[i]] == ys[j]:
            kept_positions_sorted.append(i)
            j += 1
            if j == len(xs):
                break
    kept_idx_in_input = order[np.array(kept_positions_sorted, dtype=int)]
    return xs, ys, kept_idx_in_input
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- 1) 第一阶段：保留“黄色点” ----------
def keep_more_points(x, y, x_path, y_path,
                     abs_tol=100.0, rel_tol=0.012, k_mad=2.0, q_low=15,
                     jump_min=60.0, jump_bonus=80.0, x_radius_frac=0.03,
                     max_drop_cap=220.0):
    x = np.asarray(x); y = np.asarray(y)
    x_path = np.asarray(x_path); y_path = np.asarray(y_path)
    order = np.argsort(x_path)
    x_path = x_path[order]; y_path = y_path[order]

    x_min, x_max = x_path[0], x_path[-1]
    x_clip = np.clip(x, x_min, x_max)
    y_on_path = np.interp(x_clip, x_path, y_path)

    r = y - y_on_path
    N = len(x)
    n_bins = max(20, min(120, N // 150))
    edges = np.linspace(x.min(), x.max(), n_bins + 1)
    bin_id = np.clip(np.digitize(x, edges) - 1, 0, n_bins - 1)

    thr = -(abs_tol + rel_tol * y_on_path) * np.ones_like(y, float)

    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15: 
            continue
        r_b = r[idx]
        med = np.median(r_b)
        mad = np.median(np.abs(r_b - med))
        sigma = 1.4826 * mad
        thr[idx] -= k_mad * sigma

    for b in range(n_bins):
        idx = (bin_id == b)
        if idx.sum() < 15:
            continue
        r_b = r[idx]
        thr_q = np.percentile(r_b, q_low)
        thr[idx] = np.minimum(thr[idx], thr_q)

    dy = np.diff(y_path)
    jump_idx = np.where(dy >= jump_min)[0] + 1
    jump_x = x_path[jump_idx]
    x_radius = x_radius_frac * (x.max() - x.min())
    bonus = np.zeros_like(y, float)
    for jx in jump_x:
        near = np.abs(x - jx) <= x_radius
        bonus[near] += jump_bonus
    thr -= bonus

    thr = np.maximum(thr, -max_drop_cap)
    mask_keep = r >= thr
    return mask_keep, y_on_path, thr

# ---------- 2) 可选：对特定 x 区间再放宽 ----------
def relax_regions(thr, x, regions, extra=120.0):
    # regions: [(xa, xb), ...]
    x = np.asarray(x); thr = np.asarray(thr).copy()
    for (xa, xb) in regions:
        near = (x >= xa) & (x <= xb)
        thr[near] -= extra
    return thr

# ---------- 3) 第二阶段：温和递归清理（v2） ----------
def prune_downward_recursive_v2(x, y,
                                q_hi=0.80, win=401,
                                abs_tol=90.0, rel_tol=0.020,
                                k_mad=0.8, patience=5,
                                max_rounds=10):
    x = np.asarray(x); y = np.asarray(y)
    order = np.argsort(x); xs, ys = x[order], y[order]

    def rolling_quantile(a, q, w):
        n, h = len(a), w//2
        ap = np.concatenate([np.full(h, a[0]), a, np.full(h, a[-1])])
        out = np.empty(n)
        for i in range(n):
            out[i] = np.quantile(ap[i:i+w], q)
        return out

    def rolling_mad(a, w):
        n, h = len(a), w//2
        ap = np.concatenate([np.full(h, a[0]), a, np.full(h, a[-1])])
        out = np.empty(n)
        for i in range(n):
            seg = ap[i:i+w]; med = np.median(seg)
            out[i] = 1.4826 * np.median(np.abs(seg - med))
        return out

    rounds = 0
    while rounds < max_rounds and len(ys) > 0:
        plateau = rolling_quantile(ys, q_hi, win)
        sigma   = rolling_mad(ys, win)
        floor   = plateau - (abs_tol + rel_tol*np.maximum(1.0, plateau) + k_mad*sigma)

        keep = np.ones(len(ys), bool)
        run = 0
        for i in range(len(ys)):
            if ys[i] < floor[i]:
                run += 1
            else:
                run = 0
            if run >= patience:
                keep[i] = False

        if keep.all(): break
        xs, ys = xs[keep], ys[keep]
        rounds += 1

    kept_positions_sorted, j = [], 0
    for i in range(len(order)):
        if j < len(xs) and x[order[i]] == xs[j] and y[order[i]] == ys[j]:
            kept_positions_sorted.append(i); j += 1
            if j == len(xs): break
    kept_idx_in_input = order[np.array(kept_positions_sorted, int)]
    return xs, ys, kept_idx_in_input

# ---------- 4) 一键管道 ----------
def run_pipeline(x, y, x_path, y_path,
                 preset="Generous",     # "Conservative" | "Balanced" | "Generous" | dict(自定义)
                 region_relax=None,     # [(xa, xb), ...] 需要额外放宽的区间；None=不用
                 region_extra=120.0,    # 区域额外放宽幅度
                 recursive_cfg=None,    # dict 覆盖 v2 递归的默认参数
                 plot=True):
    presets = {
        "Conservative": dict(abs_tol=60,  rel_tol=0.008, k_mad=1.5, q_low=20, jump_min=60, jump_bonus=40, x_radius_frac=0.025, max_drop_cap=160),
        "Balanced":     dict(abs_tol=100, rel_tol=0.012, k_mad=2.0, q_low=15, jump_min=60, jump_bonus=80, x_radius_frac=0.03,  max_drop_cap=220),
        "Generous":     dict(abs_tol=170, rel_tol=0.022, k_mad=3.0, q_low=8,  jump_min=45, jump_bonus=160,x_radius_frac=0.055, max_drop_cap=340),
    }
    kcfg = presets[preset] if isinstance(preset, str) else dict(preset)

    # 1) 第一阶段
    mask_keep, y_on_path, thr = keep_more_points(x, y, x_path, y_path, **kcfg)

    # 1.1) 可选：区域加权放宽（比如你红箭头所在 x 段）
    if region_relax:
        thr_relaxed = relax_regions(thr, x, region_relax, extra=region_extra)
        # 用放宽后的阈值重算 mask
        y_on_path_full = np.asarray(y_on_path)  # 与 x 等长
        r = np.asarray(y) - y_on_path_full
        mask_keep = r >= thr_relaxed

    x_keep = np.asarray(x)[mask_keep]
    y_keep = np.asarray(y)[mask_keep]

    # 2) 第二阶段：温和递归
    rcfg = dict(q_hi=0.80, win=401, abs_tol=90.0, rel_tol=0.020, k_mad=0.8, patience=5, max_rounds=20)
    if recursive_cfg:
        rcfg.update(recursive_cfg)
    x_final, y_final, kept_pos_within_keep = prune_downward_recursive_v2(x_keep, y_keep, **rcfg)

    # 最终索引映射回原始 (x,y)
    kept_idx_final = np.nonzero(mask_keep)[0][kept_pos_within_keep]

    # 3) 可视化
    if plot:
        plt.figure(figsize=(14,7))
        plt.scatter(np.asarray(x), np.asarray(y), s=6, alpha=0.06, color='gray', label='all raw points')
        plt.scatter(x_keep, y_keep, s=10, alpha=0.35, color='gold', edgecolors='none', label='kept stage-1')
        plt.scatter(x_final, y_final, s=16, alpha=0.9, color='orange', edgecolors='black', label='after recursive')
        # 路径（可选）
        op = np.argsort(x_path)
        plt.plot(np.asarray(x_path)[op], np.asarray(y_path)[op], lw=1.2, color='tomato', alpha=0.9, label='path')
        plt.xlabel('accumulative complexity'); plt.ylabel('complexity')
        plt.title('Keep-more → (optional) region relax → gentle recursive')
        plt.legend(); plt.tight_layout(); plt.show()

    return {
        "mask_stage1": mask_keep,
        "x_stage1": x_keep, "y_stage1": y_keep,
        "x_final": x_final, "y_final": y_final,
        "kept_idx_final": kept_idx_final
    }

# ---------- 5) 使用示例 ----------
# 例：把你标红的四段区域加宽（自行按你的 x 范围改）
# regions = [(0.00e6, 0.30e6), (1.35e6, 1.55e6), (2.35e6, 2.60e6), (2.85e6, 3.05e6)]
# result = run_pipeline(x, y, x_path, y_path,
#                       preset="Generous",
#                       region_relax=regions, region_extra=150.0,
#                       recursive_cfg={"q_hi":0.78, "abs_tol":110.0, "rel_tol":0.024, "patience":6},
#                       plot=True)
# 最终保留点（原始索引）：result["kept_idx_final"]
```

```python
def fit_model(X, Y, m, cv_splits=5, random_state=42, is_fast=False):
    """拟合 m 段并计算各类指标"""
    mdl = pwlf.PiecewiseLinFit(X, Y)
    if is_fast == False:
        breaks = mdl.fit(m) 
    else: 
        breaks = mdl.fastfit(m) 
    y_hat = mdl.predict(X)

    n   = len(Y)
    rss = float(np.sum((Y - y_hat)**2))
    mse = rss / n 
    mae = float(mean_absolute_error(Y, y_hat)) #Mean Absolut Error 
    tss = float(np.sum((Y - np.mean(Y))**2)) #Total Sum of Squares, TSS
    r2  = 1 - rss / tss if tss > 0 else np.nan

    # 参数个数惩罚：常用近似 k = 2*m（斜率/截距 + 断点自由度）
    k   = 2 * m
    aic = n * np.log(mse) + 2 * k #在保证模型拟合数据的同时，尽量避免模型过于复杂（overfitting）。
    bic = n * np.log(mse) + k * np.log(n) #在拟合效果和模型复杂度之间做权衡，但 比 AIC 惩罚更严格。
    r2_adj = 1 - ((rss/(n - k)) / (tss/(n - 1))) if (n - k) > 0 and (tss > 0) else np.nan

#     # 交叉验证 MSE
#     kf = KFold(n_splits=cv_splits, shuffle=True, random_state=random_state)
#     cv_mses = []
#     for tr, te in kf.split(X):
#         Xtr, Ytr = X[tr], Y[tr]
#         Xte, Yte = X[te], Y[te]
#         mdl_cv = pwlf.PiecewiseLinFit(Xtr, Ytr)
#         mdl_cv.fit(m)
#         Ypred = mdl_cv.predict(Xte)
#         cv_mses.append(mean_squared_error(Yte, Ypred))
#     cv_mse = float(np.mean(cv_mses))

    return {
        "m": m,
        "breaks": breaks,
        "train_MSE": mse,
        "train_MAE": mae,
        
        "R2": r2,
        "R2_adj": r2_adj,
        "AIC": aic,
        "BIC": bic,
#         "CV_MSE": cv_mse,
    }
```

```python
def evaluate_all_segments(X, Y, min_m=2, max_m=40, criterion="BIC",
                          cv_splits=5, random_state=42, early_stop=True, is_fast=False):
    """
    遍历分段数，返回每个 m 的指标，并根据 criterion 选最佳。
    criterion ∈ {"BIC","AIC","CV_MSE","R2_adj","MSE"}
      - MSE 越小越好（MSE 指训练集 MSE，对应键为 'train_MSE'）
      - R2_adj 越大越好
    """
    X = np.asarray(X).reshape(-1)
    Y = np.asarray(Y).reshape(-1)
    n = len(Y)

    # 基本可行性限制：至少保证每段有≥2个点（保守些）
    max_feasible = max(min(max_m, n // 2), min_m)
    ms = list(range(min_m, max_feasible + 1))

    crit = criterion.upper()
    # 将 "MSE" 统一映射到结果 dict 中的键名
    metric_key = {"MSE": "train_MSE"}.get(crit, crit)

    results = []
    for m in ms:
        t0 = time.perf_counter()
        try:
            res = fit_model(X, Y, m, cv_splits=cv_splits, random_state=random_state, is_fast=is_fast)
            t1 = time.perf_counter()
            res["time_s"] = t1 - t0
            results.append(res)
            
        except Exception as e:
            t1 = time.perf_counter()
            # 若某些 m 因数据分布问题无法拟合，记录错误并继续
            results.append({"m": m, "error": str(e), "time_s": float(t1 - t0)})

        # 提前停止邏輯 （基於所選 criterion）
        if early_stop and len(results) >= 3:
            last_three = [r for r in results[-3:] if "error" not in r and metric_key in r]
            if len(last_three) == 3:
                v0, v1, v2 = [r[metric_key] for r in last_three]
                if crit in ("MSE"):
                    # 越小越好：若連續三次指標遞增，說明開始惡化，停止
                    if (v2 > v1) and (v1 > v0):
                        break
                elif crit == "R2_ADJ":
                    # 越大越好：若连续三次指标递减，停止
                    if (v2 < v1) and (v1 < v0):
                        break

    # 选最优
    valid = [r for r in results if "error" not in r]
    if not valid:
        raise RuntimeError("没有成功的拟合结果，请检查数据或放宽可行范围。")

    if crit in ("MSE"):
        # 越小越好；NaN/缺失时丢到 +∞
        def safe_val(r):
            v = r.get(metric_key, np.inf)
            return v if (v is not None and not np.isnan(v)) else np.inf
        best = min(valid, key=safe_val)

    # 打印摘要
    print("=== Summary ===")
    for r in valid:
        print(f"m={r['m']:>2} | "
              f"MSE={r['train_MSE']:.3f} | "
              f"time={r['time_s']:.2f} s | "
              f"breaks={np.round(r['breaks'],3)}"
             
             )
        print("--------------------------")
    shown_key = metric_key 
    print(f"\n✅ Best by {criterion}: Segments(m)={best['m']}, MSE={best[shown_key]:.3f}, time:{best['time_s']:.2f}, breaks={np.round(best['breaks'], 6)}")
    return {"all": results, "best": best}
```
