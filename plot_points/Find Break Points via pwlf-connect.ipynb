```python
import pandas as pd
import pwlf
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_composer = pd.read_csv("bootstrap-cms_30segments_AME.csv")
```

```python
# Prepare the data for pwlf
xs_all = df_composer['accumulative_raw'].values
ys_all = df_composer['complexity_raw'].values
```

```python
# Fit piecewise linear model with a chosen number of segments
my_pwlf = pwlf.PiecewiseLinFit(xs_all, ys_all)
num_segments = 30
breakpoints = my_pwlf.fit(num_segments)

breakpoints
```

```python
my_pwlf.fit_with_breaks(breakpoints)
```

```python
### plot the graph 
```

```python
# --- Plot raw data ---
plt.figure(figsize=(12,5))
plt.scatter(xs_all, ys_all, s=8, label='raw')

# --- Build connected piecewise lines ---
plot_x, plot_y = [], []   # to hold a single continuous polyline
prev_a, prev_b = None, None

for i in range(len(breakpoints) - 1):
    x0, x1 = breakpoints[i], breakpoints[i+1]
    mask = (xs_all >= x0) & (xs_all <= x1)
    if mask.sum() < 2:
        continue

    Xi = xs_all[mask]
    Yi = ys_all[mask]

    # OLS fit on this segment: Yi ≈ a_raw + b * Xi
    X1 = np.vstack([np.ones_like(Xi), Xi]).T
    beta, *_ = np.linalg.lstsq(X1, Yi, rcond=None)
    a_raw, b = beta[0], beta[1]

    # From the 2nd segment on, force continuity at x0:
    # a = y_prev_at_x0 - b * x0
    if i == 0 or prev_a is None:
        a = a_raw
    else:
        y0_prev = prev_a + prev_b * x0
        a = y0_prev - b * x0

    # Generate points only within data range of this segment
    x_start = max(x0, Xi.min())
    x_end   = min(x1, Xi.max())
    if x_end <= x_start:
        continue

    xs = np.linspace(x_start, x_end, 200)
    ys = a + b * xs

    # Append to the global connected polyline (avoid duplicating the joint point)
    if len(plot_x) > 0:
        plot_x.extend(xs[1:])
        plot_y.extend(ys[1:])
    else:
        plot_x.extend(xs)
        plot_y.extend(ys)

    # Optional: also draw each segment (comment out if you only want one line)
    plt.plot(xs, ys, linewidth=2, label=f'segment {i+1}' if i == 0 else None)

    prev_a, prev_b = a, b

# --- Draw vertical lines at inner breakpoints ---
for bp in breakpoints[1:-1]:
    plt.axvline(bp, linestyle='--', alpha=0.3)

plt.title(f'Piecewise linear fit with connected segments ({num_segments} segments)')
plt.xlabel('accumulative complexity')
plt.ylabel('complexity')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
