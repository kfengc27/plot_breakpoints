```python
# === KNN 图聚类底部“水平线”并画图 ===
# 需要已有 df_composer，列名：'accumulative_raw'（x），'complexity_raw'（y）
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# ---------------- 参数（可调） ----------------
ROLLING_WINDOW = 15    # 平滑窗口
SLOPE_THRESH   = 0.02  # 近似水平的斜率阈值（标准化后）
BOTTOM_PCT     = 0.60  # 只保留 y 处于底部分位的点
K_NEIGHBORS    = 15    # kNN 中的 k（10~30 常用）
EPS_Y          = 0.12  # 在标准化后的 y 轴上，允许连边的最大 y 距离（限制“串链”）
MIN_CLUSTER_SZ = 50    # 最小簇大小（防止碎片）

# ---------------- 读取数据 ----------------
x = df_composer['accumulative_raw'].values.astype(float)
y = df_composer['complexity_raw'].values.astype(float)

# ---------------- 预处理：排序 + 平滑 + 斜率 ----------------
order = np.argsort(x)
x_sorted = x[order]
y_sorted = y[order]

# 平滑（滚动均值）
y_smooth = pd.Series(y_sorted).rolling(ROLLING_WINDOW, min_periods=1, center=True).mean().values

# 标准化（让阈值在不同数据集更稳）
x_s = (x_sorted - np.median(x_sorted)) / (np.std(x_sorted) + 1e-9)
y_s = (y_smooth - np.median(y_smooth)) / (np.std(y_smooth) + 1e-9)

# 局部斜率
dy = np.gradient(y_s)
dx = np.gradient(x_s)
slope = np.divide(dy, dx, out=np.zeros_like(dy), where=np.abs(dx) > 1e-12)

# ---------------- 选择“底部 + 近似水平”的点 ----------------
y_cut = np.percentile(y_sorted, BOTTOM_PCT * 100.0)
mask_flat = np.abs(slope) <= SLOPE_THRESH
mask_low  = y_sorted <= y_cut
mask_keep = mask_flat & mask_low

# 如果太少则放宽，仅保留底部点
if mask_keep.sum() < max(50, K_NEIGHBORS * 3):
    mask_keep = (y_sorted <= y_cut)

y_knn = y_s[mask_keep].reshape(-1, 1)  # 注意：用“标准化后的 y”建图更稳
idx_keep = np.where(mask_keep)[0]
n = len(y_knn)

# 辅助：并查集（连通分量）
parent = np.arange(n)
rank = np.zeros(n, dtype=int)

def find(u):
    while parent[u] != u:
        parent[u] = parent[parent[u]]
        u = parent[u]
    return u

def union(u, v):
    ru, rv = find(u), find(v)
    if ru == rv: 
        return
    if rank[ru] < rank[rv]:
        parent[ru] = rv
    elif rank[ru] > rank[rv]:
        parent[rv] = ru
    else:
        parent[rv] = ru
        rank[ru] += 1

# ---------------- 建立 kNN 图（互为近邻 + y 距离限制） ----------------
if n > 0:
    nbrs = NearestNeighbors(n_neighbors=min(K_NEIGHBORS, n), algorithm='auto')
    nbrs.fit(y_knn)
    distances, indices = nbrs.kneighbors(y_knn)

    # 互为近邻快速判断表
    mutual = [set(row) for row in indices]

    for i in range(n):
        for j in indices[i]:
            if i == j:
                continue
            # 互为近邻 + y 距离不超过 EPS_Y（防止“串联”多个水平带）
            if i in mutual[j] and abs(y_knn[i, 0] - y_knn[j, 0]) <= EPS_Y:
                union(i, j)

    # 统计连通分量
    from collections import defaultdict
    groups = defaultdict(list)
    for i in range(n):
        groups[find(i)].append(i)

    # 过滤太小的分量
    clusters_local = {root: idxs for root, idxs in groups.items() if len(idxs) >= MIN_CLUSTER_SZ}
else:
    clusters_local = {}

# ---------------- 映射回原始数据的簇标签 ----------------
labels_sorted = -1 * np.ones_like(y_sorted, dtype=int)
label_id = 0
for root, idxs in clusters_local.items():
    # 将局部索引映射回排序数组的索引
    global_indices = idx_keep[np.array(idxs, dtype=int)]
    labels_sorted[global_indices] = label_id
    label_id += 1

# 回到原始顺序
labels_back = -1 * np.ones_like(y, dtype=int)
labels_back[order] = labels_sorted
df_composer['knn_bottom_line'] = labels_back

# ---------------- 画图 ----------------
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=8, alpha=0.35, label='raw')

for lab in sorted(set(labels_back) - {-1}):
    sel = (labels_back == lab)
    plt.scatter(x[sel], y[sel], s=18, alpha=0.9, label=f'kNN bottom line {lab}')

plt.xlabel('accumulative complexity')
plt.ylabel('complexity')
plt.title('k-NN graph clustering of bottom horizontal lines')
plt.legend()
plt.show()

# （可选）打印每条“底线”的统计
for lab in sorted(set(labels_back) - {-1}):
    yy = y[labels_back == lab]
    print(f'Cluster {lab}: y≈{np.mean(yy):.2f} ± {np.std(yy):.2f}, n={len(yy)}')
```

```python
# This cell clusters the bottom horizontal "lines" (plateaus) using KMeans on raw points' y-values
# and plots the result. It expects a DataFrame `df_composer` with columns
# 'accumulative_raw' and 'complexity_raw'.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ---------------------
# 1) Load x, y
# ---------------------
if 'df_composer' not in globals():
    raise RuntimeError("未找到 df_composer。请先读取数据到 df_composer（包含 'accumulative_raw' 和 'complexity_raw' 列）。")

x = df_composer['accumulative_raw'].values.astype(float)
y = df_composer['complexity_raw'].values.astype(float)

# ---------------------
# 2) Parameters you can tune
# ---------------------
ROLLING_WINDOW = 15     # 平滑窗口
SLOPE_THRESH   = 0.02   # 斜率阈值（在标准化后）
BOTTOM_PCT     = 0.60   # 只保留 y 处于底部分位的点
N_CLUSTERS     = 3      # 想得到几条“底线”

# ---------------------
# 3) Prepare: sorting, smoothing, slope
# ---------------------
order = np.argsort(x)
x_sorted = x[order]
y_sorted = y[order]

# 平滑（滚动均值）
y_smooth = pd.Series(y_sorted).rolling(ROLLING_WINDOW, min_periods=1, center=True).mean().values

# 标准化以便阈值稳定
x_s = (x_sorted - np.median(x_sorted)) / (np.std(x_sorted) + 1e-9)
y_s = (y_smooth - np.median(y_smooth)) / (np.std(y_smooth) + 1e-9)

# 计算局部斜率 dy/dx
dy = np.gradient(y_s)
dx = np.gradient(x_s)
slope = np.divide(dy, dx, out=np.zeros_like(dy), where=np.abs(dx) > 1e-12)

# ---------------------
# 4) 选择“底部 + 近似水平”的点
# ---------------------
y_cut = np.percentile(y_sorted, BOTTOM_PCT * 100.0)
mask_flat  = np.abs(slope) <= SLOPE_THRESH
mask_low   = y_sorted <= y_cut
mask_keep  = mask_flat & mask_low

# 如果保留点过少，放宽条件
if mask_keep.sum() < max(50, N_CLUSTERS * 20):
    mask_keep = (y_sorted <= y_cut)

X_km = y_sorted[mask_keep].reshape(-1, 1)  # 仅用 y 值做聚类（适合水平线）
x_km = x_sorted[mask_keep]

# ---------------------
# 5) KMeans 聚类
# ---------------------
K = min(N_CLUSTERS, len(X_km)) if len(X_km) > 0 else 1
kmeans = KMeans(n_clusters=K, n_init=20, random_state=0)
labels = kmeans.fit_predict(X_km) if len(X_km) > 0 else np.array([])

# ---------------------
# 6) 将簇标签映射回原始顺序（未被选中的点设为 -1）
# ---------------------
labels_sorted = -1 * np.ones_like(y_sorted, dtype=int)
idx_keep = np.where(mask_keep)[0]
if len(labels) > 0:
    labels_sorted[idx_keep] = labels

labels_back = -1 * np.ones_like(y, dtype=int)
labels_back[order] = labels_sorted
df_composer['kmeans_bottom_line'] = labels_back

# ---------------------
# 7) 画图
# ---------------------
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=8, alpha=0.4, label='raw')

# 仅将聚类到的底部点再画一遍（不同簇会自动使用不同颜色）
if len(labels) > 0:
    # 为避免与上面的 raw 叠在一起难以区分，这里稍微增大点的尺寸与透明度
    for lab in sorted(set(labels)):
        sel = (labels_back == lab)
        plt.scatter(x[sel], y[sel], s=18, alpha=0.9, label=f'KMeans bottom line {lab}')

plt.xlabel('accumulative complexity')
plt.ylabel('complexity')
plt.title('KMeans clustering of bottom horizontal lines (raw-point method)')
plt.legend()

plt.show()

out_csv
```
